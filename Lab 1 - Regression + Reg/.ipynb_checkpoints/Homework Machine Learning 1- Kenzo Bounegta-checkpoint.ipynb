## Question 3 and 4 : we include the training error in the plots

We will define the four functions mentionned in the exercise

```python
df_without_sunspot = df.drop('Sunspot_Count',axis=1)
```

```python
def f1(x):
    j = np.arange(1, 6)
    X = x**j
    bias_var = np.ones((np.shape(x)[0], 1))
    return np.concatenate((bias_var, X), axis=1)
```

```python
def f2(x):
    j = np.arange(0, 51, 5)
    X = np.exp(-((40*x - j)**2)/25)
    bias_var = np.ones((np.shape(X)[0], 1))
    return np.concatenate((bias_var, X), axis = 1)
```

```python
def f3(x):
    j = np.arange(1, 6)
    X = np.cos(x/j)
    bias_var = np.ones((np.shape(X)[0], 1))
    return np.concatenate((bias_var, X), axis = 1)
```

```python
def f4(x):
    j = np.arange(1, 26)
    X = np.cos(x/j)
    bias_var = np.ones((np.shape(X)[0], 1))
    return np.concatenate((bias_var, X), axis = 1)
```

```python
year = df['Year'].to_numpy()[:, np.newaxis] # we use broadcasting for later computations 
rep_count = df['Republican_Count'].to_numpy()[:, np.newaxis] # we use broadcasting for later computations 
```

Function 1

```python
R = f1(year)
solver = np.linalg.solve(R.T @ R, R.T @ rep_count) # we solve x = (XTX)^1XTy .
rmse = np.sqrt(np.mean((R@solver - rep_count)**2))
train_error = 0.5*np.sum((R @ solver - rep_count)**2)

x = np.linspace(np.min(year), np.max(year), 100)[:, np.newaxis]
plt.plot(x, f1(x)@solver, label = 'f1(x)')
plt.scatter(year, rep_count, label='Real data', color='red')
plt.xlabel('Time')
plt.ylabel('Number of Republican in Senate')
plt.legend(loc="lower right")
plt.title(f"Train error is {train_error} and the RMSE is {rmse} ")
```

Function 2

```python
R = f2(year)
solver = np.linalg.solve(R.T @ R, R.T @ rep_count)  
rmse = np.sqrt(np.mean((R@solver - rep_count)**2))
train_error = 0.5*np.sum((R @ solver - rep_count)**2)

x = np.linspace(np.min(year), np.max(year), 100)[:, np.newaxis]
plt.plot(x, f2(x)@solver, label = 'f2(x)')
plt.scatter(year, rep_count, label='Real data', color='red')
plt.xlabel('Time')
plt.ylabel('Number of Republican in Senate')
plt.legend(loc="lower right")
plt.title(f"Train error is {train_error} and the RMSE is {rmse} ")
```

Function 3

```python
R = f3(year)
solver = np.linalg.solve(R.T @ R, R.T @ rep_count)  
rmse = np.sqrt(np.mean((R@solver - rep_count)**2))
train_error = 0.5*np.sum((R @ solver - rep_count)**2)

x = np.linspace(np.min(year), np.max(year), 100)[:, np.newaxis]
plt.plot(x, f3(x)@solver, label = 'f3(x)')
plt.scatter(year, rep_count, label='Real data', color='red')
plt.xlabel('Time')
plt.ylabel('Number of Republican in Senate')
plt.legend(loc="lower right")
plt.title(f"Train error is {train_error} and the RMSE is {rmse} ")
```

Function 4

```python
R = f4(year)
solver = np.linalg.solve(R.T @ R, R.T @ rep_count)  
rmse = np.sqrt(np.mean((R@solver - rep_count)**2))
train_error = 0.5*np.sum((R @ solver - rep_count)**2)

x = np.linspace(np.min(year), np.max(year), 100)[:, np.newaxis]
plt.plot(x, f4(x)@solver, label = 'f4(x)')
plt.scatter(year, rep_count, label='Real data', color='red')
plt.xlabel('Time')
plt.ylabel('Number of Republican in Senate')
plt.legend(loc="lower right")
plt.title(f"Train error is {train_error} and the RMSE is {rmse} ")
```

We first scale our data

Function 1
