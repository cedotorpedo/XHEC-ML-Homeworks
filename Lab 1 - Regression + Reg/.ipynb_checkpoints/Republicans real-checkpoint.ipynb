# Exo 2  - Scale

```python
years_scaled = (republicans["Year"] - 1960) / 40
```

## Exercice 3.1 - Feature engineering

```python
def X_features(x, basis):
    
    if basis == 'poly':
        
        X = np.zeros((x.shape[0], 6)) #we have 6 columns instead of 5 because of the bias/intercept term
        for i in range(6):
            X[:, i] = x**i   
        return X
    
    elif basis == 'expo':
        
        mu = np.arange(0, 55, 5)
        X = np.zeros((x.shape[0], len(mu)+1)) 
        X[:, 0] = np.ones(x.shape[0]) #Here again we add the first column to be the bias term
        for i in range(1, len(mu)+1):
            X[:, i] = np.exp(-(40 * x-mu[i-1])**2 / 25)
        return X
    
    elif basis == 'coss':
    
        X = np.zeros((x.shape[0], 6)) #we have 6 columns instead of 5 because of the bias/intercept term
        X[:,0] = np.ones(x.shape[0])
    
        for i in range(1, 6):
            X[:, i] = np.cos(x / i)
        return X
    
    elif basis == 'cosb':
        X = np.zeros((x.shape[0], 26)) #we have 26 columns instead of 25 because of the bias/intercept term
        X[:, 0] = np.ones(x.shape[0])
    
        for i in range(1, 26):
            X[:, i] = np.cos(x / i)
        features = X    
        return features
```

```python
def polynomial_features(x):

    X = np.zeros((x.shape[0],6)) #we have 6 columns instead of 5 because of the bias/intercept term
    for i in range(6):
        X[:,i] = x**i   
    return X
```

```python
def exponential_features(x):

    mu = np.arange(0,55,5)
    X = np.zeros((x.shape[0],len(mu)+1)) 
    X[:,0] = np.ones(x.shape[0]) #Here again we add the first column to be the bias term
    
    for i in range(1,len(mu)+1):
        X[:,i] = np.exp(-(40*x-mu[i-1])**2/25)
    return X
```

```python
def cos_features_small(x):
    
    X = np.zeros((x.shape[0],6)) #we have 6 columns instead of 5 because of the bias/intercept term
    X[:,0] = np.ones(x.shape[0])
    
    for i in range(1,6):
        X[:,i] = np.cos(x/i)
    return X
```

```python
def cos_features_big(x):
    X = np.zeros((x.shape[0],26)) #we have 26 columns instead of 25 because of the bias/intercept term
    X[:,0] = np.ones(x.shape[0])
    
    for i in range(1,26):
        X[:,i] = np.cos(x/i)
    features = X    
    return features
```

## Exercice 3.2 - fitting w

```python
def fit_least_square(X, y):
    
    w_star = np.linalg.solve(X.T@X, X.T@y)
    y_hat = X@w_star
    
    return w_star, y_hat
```

```python
X_train_poly = polynomial_features(years_scaled)
X_train_expo = exponential_features(years_scaled)
X_train_coss = cos_features_small(years_scaled)
X_train_cosb = cos_features_big(years_scaled)
```

```python
y_train_sunspot = republicans["Sunspot_Count"]
y_train_rep = republicans["Republican_Count"]
```

```python
w_s_poly, y_hat_train_s_poly = fit_least_square(X_train_poly, y_train_sunspot)
w_r_poly, y_hat_train_r_poly = fit_least_square(X_train_poly, y_train_rep)

w_s_expo, y_hat_train_s_expo = fit_least_square(X_train_expo, y_train_sunspot)
w_r_expo, y_hat_train_r_expo = fit_least_square(X_train_expo, y_train_rep)

w_s_coss, y_hat_train_s_coss = fit_least_square(X_train_coss, y_train_sunspot)
w_r_coss, y_hat_train_r_coss = fit_least_square(X_train_coss, y_train_rep)

w_s_cosb, y_hat_train_s_cosb = fit_least_square(X_train_cosb, y_train_sunspot)
w_r_cosb, y_hat_train_r_cosb = fit_least_square(X_train_cosb, y_train_rep)
```

```python
fig, ax = plt.subplots(2,figsize=(10, 6))

colors = ["k","r","g","b"]
styles = [':','--','-','dashdot']
labels = ['Poly','Expo','Cos small','Cos big']


ax[0].scatter(republicans["Year"],republicans["Sunspot_Count"])
ax[1].scatter(republicans["Year"],republicans["Republican_Count"])

ax[0].plot(republicans["Year"], y_hat_train_s_poly, color = colors[0], linestyle = styles[0], label = labels[0])
ax[0].plot(republicans["Year"], y_hat_train_s_expo, color = colors[1], linestyle = styles[1], label = labels[1])
ax[0].plot(republicans["Year"], y_hat_train_s_coss, color = colors[2], linestyle = styles[2], label = labels[2])
ax[0].plot(republicans["Year"], y_hat_train_s_cosb, color = colors[3], linestyle = styles[3], label = labels[3])

ax[0].set_title("Sunspots and their regressions",fontsize = 12)
ax[0].legend()

ax[1].plot(republicans["Year"], y_hat_train_r_poly, color = colors[0], linestyle = styles[0], label = labels[0])
ax[1].plot(republicans["Year"], y_hat_train_r_expo, color = colors[1], linestyle = styles[1], label = labels[1])
ax[1].plot(republicans["Year"], y_hat_train_r_coss, color = colors[2], linestyle = styles[2], label = labels[2])
ax[1].plot(republicans["Year"], y_hat_train_r_cosb, color = colors[3], linestyle = styles[3], label = labels[3])
ax[1].set_title("Republican spots and their regressions",fontsize = 12)
ax[1].legend()

plt.tight_layout()
plt.show()
```

## Exercice 4 - Plotting the error

```python
def error_function(X, w, y):
    N = X.shape[0]
    cost = (X@w - y)**2
    total_cost= np.sum(cost)/(2*N)
    return total_cost
```

```python
error_s_poly = error_function(X_train_poly, w_s_poly, y_train_sunspot)
error_s_expo = error_function(X_train_expo, w_s_expo, y_train_sunspot)
error_s_coss = error_function(X_train_coss, w_s_coss, y_train_sunspot)
error_s_cosb = error_function(X_train_cosb, w_s_cosb, y_train_sunspot)

error_r_poly = error_function(X_train_poly, w_r_poly, y_train_rep)
error_r_expo = error_function(X_train_expo, w_r_expo, y_train_rep)
error_r_coss = error_function(X_train_coss, w_r_coss, y_train_rep)
error_r_cosb = error_function(X_train_cosb, w_r_cosb, y_train_rep)
```

```python
names = ['Poly', 'Expo', 'Cos small','Cos big']
errors_sunspot_regression = [error_s_poly, error_s_expo, error_s_coss, error_s_cosb]
errors_republican_regression = [error_r_poly, error_r_expo, error_r_coss, error_r_cosb]

fig, ax = plt.subplots(2,figsize=(9, 5))

ax[0].bar(names, errors_sunspot_regression)

ax[1].bar(names, errors_republican_regression)
```

## Exercice 5 - Sunspots vs Republicans

Repeat the same exact process as above but for Number of Sunspots (x-axis) v. Number
of Republicans in the Senate (y-axis). Here, to avoid numerical instability with numbers
in the Sunspot Count column, we will also scale the data first by dividing all “sunspot count”
inputs by 20. In addition, only use data from before 1985, and only use basis functions (a),
(c), and (d) – ignore basis (b). You will therefore make 3 total plots. For each plot make
sure to also include the train error

```python
sunspot = republicans["Sunspot_Count"][republicans["Year"]<1985]
sunspot_scaled = sunspot/20

y_train_rep = y_train_rep[republicans["Year"]<1985]
```

```python
X_train_poly = polynomial_features(sunspot_scaled)
X_train_coss = cos_features_small(sunspot_scaled)
X_train_cosb = cos_features_big(sunspot_scaled)

w_poly, y_hat_train_poly = fit_least_square(X_train_poly, y_train_rep)
w_coss, y_hat_train_coss = fit_least_square(X_train_coss, y_train_rep)
w_cosb, y_hat_train_cosb = fit_least_square(X_train_cosb, y_train_rep)
```

```python
sunspot_sorted = np.sort(sunspot)
indexes_sorted = np.argsort(sunspot).to_numpy()

rep_sorted = [y_train_rep[idx] for idx in indexes_sorted]
y_hat_poly_sorted = [y_hat_train_poly[idx] for idx in indexes_sorted]
y_hat_coss_sorted = [y_hat_train_coss[idx] for idx in indexes_sorted]
y_hat_cosb_sorted = [y_hat_train_cosb[idx] for idx in indexes_sorted]
```

```python
fig, ax = plt.subplots(1,figsize=(10, 6))

colors = ["k","r","g"]
styles = [':','--','-']
labels = ['Poly','Cos small','Cos big']

ax.scatter(sunspot,y_train_rep)
ax.plot(sunspot_sorted,y_hat_poly_sorted, color = colors[0], linestyle = styles[0], label = labels[0])
ax.plot(sunspot_sorted,y_hat_coss_sorted, color = colors[1], linestyle = styles[1], label = labels[1])
ax.plot(sunspot_sorted,y_hat_cosb_sorted, color = colors[2], linestyle = styles[2], label = labels[2])
ax.set_title("Republicans in senate vs sunspots",fontsize = 12)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
error_poly = error_function(X_train_poly, w_poly, y_train_rep)
error_coss = error_function(X_train_coss, w_coss, y_train_rep)
error_cosb = error_function(X_train_cosb, w_cosb, y_train_rep)
```

```python
names = ['Poly', 'Cos small','Cos big']
errors_sunspot_regression = [error_poly, error_coss, error_cosb]

fig, ax = plt.subplots(1,figsize=(9, 5))

ax.bar(names, errors_sunspot_regression)
```
