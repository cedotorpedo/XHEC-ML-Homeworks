# Exercice 3
## 1 - Feature engineering

First we write a function that when

```python
def X_features(x, basis):
    
    """
    This function allows us to create different design matrices based on the selected
    basis function and the initial feature vector x.
    """
    
    if basis == 'poly':
        
        X = np.zeros((x.shape[0], 6)) #we have 6 columns instead of 5 because of the bias/intercept term
        for i in range(6):
            X[:, i] = x**i   
        return X
    
    elif basis == 'expo':
        
        mu = np.arange(0, 55, 5)
        X = np.zeros((x.shape[0], len(mu)+1)) 
        X[:, 0] = np.ones(x.shape[0]) #Here again we add the first column to be the bias term
        for i in range(1, len(mu)+1):
            X[:, i] = np.exp(-(40 * x-mu[i-1])**2 / 25)
        return X
    
    elif basis == 'coss':
    
        X = np.zeros((x.shape[0], 6)) #we have 6 columns instead of 5 because of the bias/intercept term
        X[:,0] = np.ones(x.shape[0])
    
        for i in range(1, 6):
            X[:, i] = np.cos(x / i)
        return X
    
    elif basis == 'cosb':
        X = np.zeros((x.shape[0], 26)) #we have 26 columns instead of 25 because of the bias/intercept term
        X[:, 0] = np.ones(x.shape[0])
    
        for i in range(1, 26):
            X[:, i] = np.cos(x / i)
        features = X    
        return features
```

## 2 - fitting

From class we know that the least squares estimator for regression is  
$\hat{w} = (X^TX)^{-1}X^Ty$

```python
def fit_least_square(X, y):
    
    w_star = np.linalg.solve(X.T@X,X.T@y)
    y_hat = X@w_star
    
    return w_star, y_hat
```

## 3 - plotting & Exercice 4 - train error

```python
y = republicans["Republican_Count"]
```

```python
def error_function(X, w, y):
    N = X.shape[0]
    cost = (X@w - y)**2
    total_cost= np.sum(cost) / (2*N)
    return total_cost
```

```python
fig, ax = plt.subplots(2, figsize=(12, 8))

colors = ["k", "r", "g", "b"]
styles = [':', '--', '-', 'dashdot']
labels = ['Poly', 'Expo', 'Cos small', 'Cos big']
basis = ['poly', 'expo', 'coss', 'cosb']
errors = []

ax[0].scatter(republicans["Year"], y)

for c, s, l, base in zip(colors, styles, labels, basis):
    
    X_train = X_features(years_scaled, base)
    w, y_pred = fit_least_square(X_train, y)
    ax[0].plot(republicans["Year"], y_pred, color = c, linestyle = s, label = l)
    
    error = error_function(X_train, w, y)
    errors.append(error)

ax[0].set_title("Republican senate spots regressed over the calendar year in different bases",fontsize = 12)
ax[0].legend()

ax[1].set_title("Train errors for each type of regression",fontsize = 12)
ax[1].bar(labels, errors)

plt.tight_layout()
```

## Exercice 5 - Sunspots vs Republicans

Repeat the same exact process as above but for Number of Sunspots (x-axis) v. Number
of Republicans in the Senate (y-axis). Here, to avoid numerical instability with numbers
in the Sunspot Count column, we will also scale the data first by dividing all “sunspot count”
inputs by 20. In addition, only use data from before 1985, and only use basis functions (a),
(c), and (d) – ignore basis (b). You will therefore make 3 total plots. For each plot make
sure to also include the train error

```python
sunspots = republicans["Sunspot_Count"][republicans["Year"]<1985]
sunspots_scaled = sunspots/20
y_train = y[republicans["Year"]<1985]

# In the next two lines we sort the sunspots vector by increasing value
# and we keep the indexes_sorted array so that we can order the y_train vector in the same order
#later on.
sunspots_sorted = np.sort(sunspots)
indexes_sorted = np.argsort(sunspots).to_numpy()
```

```python
fig, ax = plt.subplots(2, figsize=(12, 8))

colors = ['k', 'r', 'g']
styles = [':', '--', '-']
labels = ['Poly', 'Cos small', 'Cos big']
basis = ['poly', 'coss', 'cosb']
errors = []

ax[0].scatter(sunspots, y_train)

for c, s, l, base in zip(colors, styles, labels, basis):
    
    X_train = X_features(sunspots_scaled, base)
    w, y_pred = fit_least_square(X_train, y_train)
    y_pred_sorted = [y_pred[idx] for idx in indexes_sorted]
    ax[0].plot(sunspots_sorted, y_pred_sorted, color = c, linestyle = s, label = l)
    
    error = error_function(X_train, w, y_train)
    errors.append(error)

ax[0].set_title("Republican senate spots regressed over the number of sunspots in different bases",fontsize = 12)
ax[0].legend()

ax[1].set_title("Regression train errors for each base",fontsize = 12)
ax[1].bar(labels, errors)

plt.tight_layout()
plt.show()
```

## 6
None of these three bases provides a really good fit. There is no clear choice between the different bases, **so let's look at the test error with y_test taking values after 1985, in order to determine which one generalizes best.**

```python
sunspots_test = republicans["Sunspot_Count"][republicans["Year"]>1985]
sunspots_test_scaled = sunspots_test/20
y_test = y[republicans["Year"]>1985]

#sunspots_test_sorted = np.sort(sunspots_test)
#indexes_test_sorted = np.argsort(sunspots_test).to_numpy()
```

```python
basis = ['poly', 'coss', 'cosb']
n = len(sunspots_scaled)
errors_test = []
for base in basis:
    X_train = X_features(sunspots_scaled, base)
    X_test = X_features(sunspots_test_scaled, base)
    w, _ = fit_least_square(X_train, y_train)
    y_test_pred = X_test@w
     
    errors_individual = (y_test_pred - y_test)**2
    total_error = np.sum(errors_individual) / (2*n)
    errors_test.append(total_error)
    
fig, ax = plt.subplots(1, figsize=(8, 8))    
ax.bar(labels, errors_test)
ax.set_title("Regression test errors for each base", fontsize = 12)
```

**Now we can say that the polynomial basis provides the "best" generalization, even though it still has a lot of bias and a lot of variance, as we can see the test error is significantly higher than the train error**.
