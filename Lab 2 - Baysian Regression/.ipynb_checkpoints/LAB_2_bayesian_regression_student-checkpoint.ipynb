## Question 2
> Sample $y$ according to the Bayesian model described previously.
Plot the on the same figure:
>1. the graph of the function $x' \in [0, 2\pi] \mapsto \phi(x')w$;
>1. the observations $y_1, \dots, y_n$ against $x_1, \dots, x_n$.

```python
# Define the function phi
def phi_a(x, max_deg): 
    x = np.array([x ** i for i in range(1, max_deg + 1)])
    return np.vstack((np.ones(len(x.transpose())), x))

x_train = phi_a(year, 5)

train_plot(x_train, y, 'Scaled year', 'Linear Regression for $\phi_j(x)=\sum a_jx_j$')
```

(b) $φ_j (x) = exp −(40x − µ_j )^2/25$   for   $µ_j = 0, 5, 10, . . . , 50 $


```python
# Define the function phi
def phi_b(x, mean_array): 
    x= np.array([np.exp(-((40*x - mu)**2)/25) for mu in mean_array])
    return np.row_stack((np.ones(len(x.transpose())), x))


# Get the trained data for this phi function
x_train = phi_b(year, np.arange(0, 51, 5))

train_plot(x_train, y, 'Scaled year', 'Linear Regression for $\phi_j(x)=exp(-(40x-\mu)^2/25)$')
```


(c) $φ_j (x) = cos(x/j) $ for j = 1, . . . , 5

```python
# Define the function phi
def phi_c(x, max_j): 
    x = np.array([np.cos(x/j) for j in range(1, max_j+1)])
    return np.row_stack((np.ones(len(x.transpose())), x))

# Get the trained data for this phi function
x_train = phi_c(year, 5)

train_plot(x_train, y, 'Scaled year', 'Linear Regression for $\phi_j(x)=cos(x/j)_{j\in [1,5]}$')
```

## Question 3
>Compute the Bayesian estimator/predictor $x_{\mathrm{new}} \mapsto \hat{y}^{\mathcal{D}}(x_{\mathrm{new}})$.

$\hat{y}^{\mathcal{D}}(x_{\mathrm{new}}) = \phi(x_{new})^T \beta .
S^{-1}\phi(x_{new})y$

```python
def baysian_estim(x_new):
    S = np.diag(np.ones(len(x_new))) + beta * phi_a(x_new).transpose() \
    @ phi_a(x_new)
    S_inv = np.linalg.solve(S, phi_a(x_new))
    return phi_a(x_new).transpose() @ (beta * S_inv) @ y
```
